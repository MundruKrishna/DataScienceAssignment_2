```python
import os
import pandas as pd
import random as rd
```

##### Read in the movie data: `pd.read_table`

```python
def get_movie_data():
    
    '''
    Function to read in movie related data
    
    Parameters
    ----------
    None
    
    Returns
    -------
    users: pd.DataFrame containing user data
    ratings: pd.DataFrame containing rating data
    movies: pd.DataFrame containing movie data
    '''
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_csv(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames)
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_csv(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames)
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_csv(os.path.join('../data', 'movies.dat'), 
                           sep='::', engine='python',header=None, names=mnames)

    return users, ratings, movies
```

```python
users, ratings, movies = get_movie_data()
```

```python
users.head()
```

```python
ratings.head()
```

```python
movies.head()
```

##### Clean up the `movies`

- Get the `year`
- Shorten the `title`


*Let us calculate mean rating and total number of ratings which helps us to find how movies are being rated*

```python

# Adding two rows for mean and counting the ratings given

movies["mean_rating"] = ""
movies["count_rating"] = ""
for i in range(3883):
    row1=movies.movie_id.iloc[i]
    condn = ratings["movie_id"].astype(int) == row1
    #calculating the mean
    mulcount = ratings[condn]["rating"].mean()
    movies["mean_rating"][i] = round(mulcount, 2)
    movies["count_rating"][i] = ratings[condn]["rating"].count()


movies.head()
```

*Lets us extract the year of the movies from the Movies data set and add it into the table and also add only title of the movie into an another row as only title*

```python
tmp = movies.title.str.extract('(.*) \(([0-9]+)\)')
tmp.apply(lambda x:x[0] if len(x) > 0 else None)
tmp.apply(lambda x: x[0][:40] if len(x) > 0 else None)
```

*Now let us assign the first part of the full title which is year*
*Second part can be assigned to only title*

```python
movies['year'] = tmp[1]
movies['titleonly'] = tmp[0]
```

```python
movies.head()
```

##### Join the tables with `pd.merge` (1 pts)

*Let us join the two tables first which are users and movies*

```python
movieRatings =  pd.merge(users, ratings, how='left', on='user_id') # left join with user_id for fetching ratings of respective user
movieRatings
```

*Let us again join the two tables the one which is already merged and the user details table*

```python
UserRatingsMovies=pd.merge(movieRatings, movies, how='left', on='movie_id') # left join on movie_id for fetching all the movie details with respect to users and their ratings
UserRatingsMovies
```

##### Task 2 : Now lets find the highest rated move 

*By using sort function we will be able to get the most liked movie and found as shown below*

```python
# The Data is grouped by movie_id, movie title
# Used aggregate sum function and printed the highest rated movie
finalData = UserRatingsMovies[['movie_id', 'titleonly', 'rating']].groupby(['movie_id', 'titleonly'], as_index=True).agg(['sum', 'mean'])
finalData.columns = finalData.columns.droplevel(0) 
finalData = finalData.reset_index().sort_values('sum', ascending=False)
# drop reset index created when grouping data
finalData
```

```python
print(f"Most Rated movie by all users is-- {finalData.iloc[1]['titleonly']}")
```

##### Task 3 : Best movie suggestion for couple

*Here We are taking two users randomly using randint and getting all their data which is already calculated. This data again can be used to find the most liked movie by both of them*

```python
#randomly generating user id
indiv1=rd.randint(1, 200)
temp1 = UserRatingsMovies["user_id"].astype(int) == indiv1
rdInd1 = UserRatingsMovies[temp1]
rdInd1 = rdInd1.sort_values(by = ['mean_rating', 'mean_rating'])
rdInd1.tail()
```

```python
#randomly generating user id
indiv2=rd.randint(500, 1000)
temp2 = UserRatingsMovies["user_id"].astype(int) == indiv2
rdInd2 = UserRatingsMovies[temp2]
rdInd2 = rdInd2.sort_values(by = ['mean_rating', 'count_rating'])
rdInd2.tail()
```

*Lets merge the data of the two users with innner join using common data variables*

```python

mutualintrestdata = rdInd1.merge(rdInd2, on=["movie_id", "titleonly", "mean_rating", "count_rating"], how='inner')
mutualintrestdata = mutualintrestdata.sort_values(by=['mean_rating','count_rating'])

mutualintrestdata.head()
```

*Now most liked movie by both individuals can be found with the above combined data with the help of index integers values through the reference with short title*

```python
 print(f"Best date movie suggested according to the users :  {mutualintrestdata.iloc[-1]['titleonly']} ")
```
